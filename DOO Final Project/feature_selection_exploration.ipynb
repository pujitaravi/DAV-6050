# imdb_score is our target valuable. We will change imdb_score to categorical data,and spliting it to high part and low part through median.

```python
reg = data['imdb_score'].values.tolist()
```

```python
reg.sort()
mid = int(len(reg) / 2)
if len(reg) % 2 == 0:
    median = (reg[mid-1] + reg[mid]) / 2.0
else:
    median = reg[mid]
print("The median of CFR is",median)
```

```python
def func(x):
    if x < 6.8*median:
        return "low"
    elif 6.8*median<=x:
        return "high"   

data['imdb_score_level'] = data['imdb_score'].apply(func)
```

```python
data.isnull().sum()
```

```python
data=data.drop(['imdb_score'],axis=1)
```

```python
data['imdb_score_level']=data['imdb_score_level'].replace({'low': 0,'high':1})
```

```python
data.dtypes
```

```python
data['rating']=data['rating'].replace({'R/TV-MA': 0,'PG-13/TV-14':1,'PG/TV-PG':2,'G/TV-G':3})
```

# Built-in Feature Importance

```python
X = data.drop('imdb_score_level',1)   #Remove Target Variable to Get Feature Matrix
y = data['imdb_score_level']
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor ()
model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances 
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(25).plot(kind='barh')
plt.show()
```

Ne feature has high correlation with imdb_score and high importance. I choose all feature except rotten_score because the function is similar with our target valuable imdb_score.
