# Data Cleaning Process

This file outlines the data cleaning process for our data. We got our dataset from data.world: https://data.world/hunterkempf/hbo-content-as-of-june-28-2020/workspace/file?filename=HBO_Content.csv

```python
# import packages
import pandas as pd
import numpy as np

# import data from github
data = pd.read_csv("https://raw.githubusercontent.com/YihangSun123/DAV6150/main/HBO_Content%20(1).csv")
data
```

```python
# importing packages
import missingno as msno
import matplotlib.pyplot as plt

# plotting missing value graphs
msno.bar(data)
msno.matrix(data)
```

```python
# plotting missing value heatmaps
msno.heatmap(data)
msno.dendrogram(data)
```

```python
# check for cuplicated values
data.duplicated().sum()
```

```python
# crop duplicated values
data=data.drop_duplicates()
```

```python
# resetting indeces
data=data.reset_index()
data
```

```python
# dropping index
data=data.drop(['index'],axis=1)
data
```

```python
# isolating scores into separate dataframe
data1=data[['imdb_score','rotten_score']]
```

```python
# isolating genre and platform columns
genre = data.iloc[:, 8:34]
platform = data.iloc[:, 35:]
```

```python
data['genre'] = None

genre_type = []

for i in range(data.shape[0]):
    genre_type = []
    for column in genre.columns:
        if genre.loc[i][column] == 1:
            genre_type.append(column)
    data['genre'][i] = genre_type
    
data
```

```python
data['platform'] = None

platform_type = []

for i in range(data.shape[0]):
    platform_type = []
    for column in platform.columns:
        if platform.loc[i][column] == 1:
            platform_type.append(column)
    data['platform'][i] = platform_type
    
data
```
